```python
from pyspark.sql.types import *
teams_schema=StructType([StructField('Country',StringType(),True),
                         StructField('Discipline',StringType(),True),
                         StructField('NOC',StringType(),True),
                         StructField('Event',StringType(),True)

])
```

```python
teams_raw=spark.read\
    .schema(teams_schema)\
        .option('header','true')\
    .csv('dbfs:/mnt/olympicstokyoadlssa/raw/Dataset/Teams.csv')
display(teams_raw)
```

```python
teams_df1=teams_raw.drop('Country')
display(teams_df1)
```

```python
teams_remove_accents_df=remove_accents_df(teams_df1)
display(teams_remove_accents_df)
```

```python
display(teams_remove_accents_df.groupBy('Discipline').count().orderBy('count',ascending=False))
```
