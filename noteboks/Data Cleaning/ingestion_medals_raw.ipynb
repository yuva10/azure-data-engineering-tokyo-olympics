```python
from pyspark.sql.types import *
coaches_schema=StructType([
    StructField('Rank',IntegerType(),True),
    StructField('NOC',StringType(),True),
    StructField('Gold',IntegerType(),True),
    StructField('Sliver',IntegerType(),True),
    StructField('Bronze',IntegerType(),True),
    StructField('Total',IntegerType(),True),
    StructField('Rank_by_total',IntegerType(),True)
])
```

```python
medals_raw=spark.read\
    .schema(coaches_schema)\
    .csv('dbfs:/mnt/olympicstokyoadlssa/raw/Dataset/Medals.csv',header=True)
display(medals_raw)
```

```python
display(medals_raw.orderBy('Total',ascending=False))
```
