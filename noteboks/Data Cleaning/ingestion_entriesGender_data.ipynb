```python
from pyspark.sql.types import *
entriesGender_schema=StructType([
    StructField('Discipline',StringType(),True),
    StructField('Female',IntegerType(),True),
    StructField('Male',IntegerType(),True),
    StructField('Total',IntegerType(),True)
])
```

```python
entriesGender_raw=spark.read\
    .schema(entriesGender_schema)\
    .csv('dbfs:/mnt/olympicstokyoadlssa/raw/Dataset/EntriesGender.csv',header=True)
display(entriesGender_raw)
```

```python
from pyspark.sql.functions import sum as _sum; 
display(entriesGender_raw.agg(_sum('Total')).collect())
```
